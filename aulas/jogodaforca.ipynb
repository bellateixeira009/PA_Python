```python
import random

def escolher_palavra():
    palavras = ["caneta", "cachorro", "arroz", "tecnologia", "arvore", "livro", "feijao", "gato"]
    return random.choice(palavras)

def mostrar_palavra(palavra, letras_certas):
    exibicao = ""
    for letra in palavra:
        if letra in letras_certas:
            exibicao += letra + " "
        else:
            exibicao += "_ "
    return exibicao.strip()

def tentar_letra(letra, palavra, letras_certas, letras_erradas):
    if letra in palavra:
        if letra not in letras_certas:
            letras_certas.append(letra)
        print(f"Boa! A letra '{letra}' está na palavra.")
    else:
        if letra not in letras_erradas:
            letras_erradas.append(letra)
            print(f"Errou! A letra '{letra}' não existe na palavra.")

def verificar_vitoria(palavra, letras_certas):
    for letra in palavra:
        if letra not in letras_certas:
            return False
    return True

def jogo_da_forca():
    palavra = escolher_palavra()
    letras_certas = []
    letras_erradas = []
    tentativas = 6

    print("=== JOGO DA FORCA ===")

    while True:
        print("\nPalavra:", mostrar_palavra(palavra, letras_certas))
        print("Letras erradas:", " ".join(letras_erradas))
        print(f"Tentativas restantes: {tentativas - len(letras_erradas)}")

        letra = input("Digite uma letra: ").lower()

        if len(letra) != 1 or not letra.isalpha():
            print("Digite apenas uma letra válida.")
            continue

        tentar_letra(letra, palavra, letras_certas, letras_erradas)

        if verificar_vitoria(palavra, letras_certas):
            print("\nVocê venceu! A palavra era:", palavra)
            break

        if len(letras_erradas) >= tentativas:
            print("\nVocê perdeu! A palavra era:", palavra)
            break



jogo_da_forca()


```
